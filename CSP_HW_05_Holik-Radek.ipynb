## Task 2
**(5 points)** Let $Y_1, \dots, Y_n$ be a random sample from normal distribution with unknown $\mu$ and known $\sigma$. Show that $\overline{Y}$ is a sufficient statistic for $\mu$.

The  likelihood function:<p>
$ L(y_1, \ldots, y_n|\mu) = (2\pi)^{-n/2} \sigma^{-n} exp\left(\frac{-1}{2 \sigma^2} \Sigma_{i=1}^n (y_i - \mu)^2 \right)$<p>

$\sigma$ is known:<p>
$\sigma^2 = \Sigma_{i=1}^n (y_i - \mu)^2 = \Sigma_{i=1}^n y_i^2 - 2 n \frac{1}{n} \Sigma_{i=1}^n y_i \mu + \Sigma_{i=1}^n \mu^2 =\Sigma_{i=1}^n y_i^2 - 2 n \bar{y} \mu + n\mu^2$<p>
Then $L$:<p>
$ L(y_1, \ldots, y_n|\mu) = exp \left(- \frac{1}{2 \sigma^2} (-2 n \bar{y} \mu + n \mu^2) \right)  (2\pi)^{-n/2} \sigma^{-n} exp\left(- \frac{1}{2 \sigma^2} \Sigma_{i=1}^n y_i^2 \right)$ <p>
This is the Factorization theorem:<p>
$L(y_1, \ldots, y_n|\mu) = g(u, \mu) h(y_1, \dots , y_n)$<p>
Where:<p>
$g(u, \mu) =  exp \left(- \frac{1}{2 \sigma^2} (-2 n \bar{y} \mu + n \mu^2) \right)$<p>
$h(y_1, \ldots, y_n) = (2\pi)^{-n/2} \sigma^{-n} exp\left(- \frac{1}{2 \sigma^2} \Sigma_{i=1}^n y_i^2 \right)$<p>
### The factorization criterion shows that $\overline{Y}$ is a sufficient statistic for $\mu$.

The Pareto distribution of parameters $\alpha, \beta > 0$ is defined by the pdf
#### <center>$ f(y) = \begin{cases} \alpha \beta^{\alpha} y^{-(\alpha + 1)} \text{, if }  y \geq \beta \\ 0 \hspace{4em} \text{ otherwise} \end{cases}  $</center>

(The Pareto distribution is often used in economics to model income distributions.)<p>
If $\beta$ is known, show that $\Pi_{i=1}^n Y_i$ is a sufficient statistic for $\alpha$.

The  likelihood function:<p>
$ L(y_1, \dots, y_n|\alpha) = \Pi_{i=1}^n \alpha \beta y_i^{-(\alpha + 1)}$, where $y_i \geq \beta $ <p>

$ = \alpha \beta y_1^{-(\alpha + 1)} \cdot \ldots \cdot \alpha \beta y_n^{-(\alpha + 1)} = \alpha^n \beta^n \frac{1}{y_1^{\alpha} y_1} \cdot \ldots \cdot \frac{1}{y_n^{\alpha}y_n}$<p>

Then $L$:<p>
$ L(y_1, \ldots, y_n|\alpha) = \alpha^n (y_1 \cdot \ldots \cdot y_n)^{-\alpha} \cdot \beta^n (y_1 \cdot \ldots \cdot y_n)^{-1}$ <p>
This is the Factorization theorem:<p>
$L(y_1, \ldots, y_n|\alpha) = g(u, \alpha) h(y_1, \ldots , y_n)$<p>
Where:<p>
$g(u, \alpha) =  \alpha^n (y_1 \cdot \ldots \cdot y_n)^{-\alpha}$<p>
$h(y_1, \ldots, y_n) = \beta^n (y_1 \cdot \ldots \cdot y_n)^{-1}$<p>
### The factorization criterion shows that $\Pi_{i=1}^n Y_i$ is a sufficient statistic for $\alpha$.

Let $X$ denote the proportion of allotted time that a randomly selected student spends working on a certain aptitude test. Suppose the pdf of $X$ is<p>
#### <center>$f_{\theta}(x) = (\theta + 1)x^{\theta} \text{ for } \theta < x < 1 \text{ and } 0 \text{ otherwise,}$</center><p>
where $\theta > 1$ is an unknown parameter. A random sample of $10$ students yields data $x_1 = .92$, $x_2 = .79$, $x_3 = .90$, $x_4 = .65$, $x_5 = .86$, $x_6 = .47$, $x_7 = .73$, $x_8 = .97$, $x_9 = .84$, $x_{10} = .67$.

```python
import numpy as np
```

```python
sampleX = [.92, .79, .90, .65, .86, .47, .73, .97, .84, .67]
```

### **(a)**
Use the method of moments to obtain an estimator of $\theta$, and then compute the estimate for this sample.

Because we are looking for only one unknown parameter, therefore calculate the first moment.<p>
$E(X) = \bar{x} = \int_0^1 x (\theta + 1) x^{\theta} dx = (\theta + 1) \int_0^1 x^{\theta + 1} dx = \frac{\theta + 1}{\theta + 2}\left[ x^{\theta + 2} \right]_0^1 = \frac{\theta + 1}{\theta + 2}$<p>
Then:<p>
$\bar{x} = \frac{\theta + 1}{\theta + 2}$<p>
$(\theta + 2) \bar{x} = \theta  + 1$<p>
$\theta \bar{x} + 2 \bar{x} = \theta + 1$<p>
$\theta \bar{x} - \theta = 1 - 2 \bar{x}$<p>
$\theta (\bar{x} - 1) = 1 - 2\bar{x}$<p>
$\theta = \frac{1 - 2\bar{x}}{\bar{x} - 1} = \frac{2\bar{x} - 1 }{1 - \bar{x}} = \frac{-2(1 - \bar{x} ) + 1}{1 - \bar{x}} = \frac{1}{1 - \bar{x}} - 2$<p>
Then we get the estimator:<p>
### <center> $\hat{\theta} = \frac{1}{ 1 - \bar{x}} -2 $ </center>


#### Estimates for the sample:

- $\bar{x}$

```python
meanX = np.mean(sampleX)
print(f"x = {meanX}")
```

- $\hat{\theta}$

```python
thetaHat1 = 1 / (1 - meanX) - 2
print(f"{thetaHat1}")
```

### **(b)**
Compute the maximum likelihood estimator of $\theta$, and then compute the estimate for this data.

First of all, we will express the likelijood function:<p>
$L(x_1, \ldots , x_n | \theta) = \Pi_{i=1}^n (\theta + 1) x_i^{\theta} = (\theta + 1)^n (x_1^{\theta} \cdot \ldots \cdot x_n^{\theta}) = (\theta + 1 )^n (x_1 \cdot \ldots \cdot x_n)^{\theta}$<p>
Now we look for $ln(L)$<p>
$ln(L(x_1, \ldots , x_n | \theta)) = n \cdot ln(\theta + 1) + \theta \cdot ln(x_1 \cdot \ldots \cdot x_n)$<p>
We want to get the maximum of $ln(L)$ due to it being the maximum of $L$.<p>
The following derivations must apply:<p>
$\frac{d ln(L)}{d \theta} = 0 \text{ and } \frac{d^2 ln(L)}{d \theta^2} < 0$<p>
Then:<p>
$\frac{d ln(L)}{d \theta} = \frac{n}{\theta + 1} + ln(x_1 \cdot \ldots \cdot x_n)$<p>
$\frac{n}{\theta + 1} + ln(x_1 \cdot \ldots \cdot x_n) = 0$<p>
$\frac{n}{\theta + 1} = - ln(x_1 \cdot \ldots \cdot x_n)$<p>
$\theta = - \frac{n}{ln(x_1 \cdot \ldots \cdot x_n)} - 1 = \frac{- n}{\Sigma_{i=1}^n ln(x_i)} - 1$<p>
We have to test the second derivative:<p>
$\frac{d^2 ln(L)}{d \theta^2} = - n \cdot (\theta + 1)^{-2} = \frac{-n}{(\theta + 1)^2} < 0 \text{ , } \forall \theta > -1  $<p>
Then we get the maximum likelihood estimator:<p>
### <center> $\hat{\theta} = \frac{-n}{\Sigma_{i=1}^n ln(x_i)} - 1 $ </center>


#### Estimates for the sample:

- $\hat{\theta}$

```python
thetaHat2 = - len(sampleX) / np.sum(np.log(sampleX)) - 1
print(f"{thetaHat2}")
```

### The estimate for the sample is approximately $2.72982$.

## Task 5 
**(5 points)**
Let $X_1, \ldots ,X_{10}$ be a random sample from a uniform distribution on the interval $[0, \theta]$. Find the likelihood function $L(x_1, \ldots, x_{10}; )$. Hint: you may wish to distinguish two cases: when $\theta < max\{x_1, \ldots, x_{10}\}$ and when $\theta \geq max\{x_1, \ldots, x_{10}\}$.
Show that the maximum likelihood estimator of $\theta$ is $max\{X_1, \ldots, X_{10}\}$.

First, we have to express pdf:<p>
#### <center>$ f(x_i; \theta)_{1 \leq i\leq 10} = \begin{cases} \frac{1}{\theta} \text{ , if }  0 \leq x_i \leq \theta  \\ 0 \text{ , if } x_i < 0 \lor x_i > \theta \end{cases}  $</center><p>
Then the likelihood function is:<p>
### <center> $L(x_1, \ldots , x_{10}| \theta ) = \begin{cases} \Pi_{i=1}^{10} \frac{1}{\theta} \hspace{0.6em} \text{ , if }  \theta \geq max\{x_1, \dots, x_{10} \}  \\ 0 \hspace{2.6em} \text{ , if } \theta < max\{x_1, \dots, x_{10} \} \end{cases}$ </center>

```python
import random
import matplotlib.pyplot as plt
```

```python
def LH(theta, x):
    if theta >= np.max(x):
        return (1 / theta)**(len(x))
    else:
        return 0
    
def estimator(x):
    return np.max(x)
```

```python
x =  [random.random() for i in range(10)]
theta = np.linspace(0.0, 2, num=1000)

mle = []
for t in theta:
    mle.append(LH(t, x))
    
thetaHat = estimator(x)
```

```python
# Set up the figure for plotting
plt.figure(figsize=(12, 8))

# plotting the data
plt.plot(theta, mle, color='blue')

# plotting the estimate
plt.scatter(thetaHat, np.max(mle), color='red', lw=5)

plt.scatter(x, np.zeros(len(x)), color='limegreen', lw=5)
 
# Adding the title
plt.title(r"Likelihood Function : $L(x_1, \ldots , x_{10}| \theta )$", fontsize=20)
 
# Adding the labels
# plt.ylabel("y-axis")
plt.xlabel(r"$\theta$", fontsize=20)

# Adding the legends
plt.legend([r"$L(x_1, \ldots, x_{10}| \theta)$", "Maximum Likelihood",r"Samples: $x_1, \ldots, x_{10}$"], fontsize=16)

plt.show()
```

From the above graph, it can be concluded that the $max\{x_1, \dots , x_{10} \}$ is also the maximum of likelihood function and thus the estimate of the $\theta$ parameter.<p>
Then the following is true:<p>
### <center> $ \hat{\theta} = max\{X_1, \ldots, X_{10}\} $ </center>
